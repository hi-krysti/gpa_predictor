```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('sat_gpa_v1.csv')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
sns.lmplot(y='sat_sum', x='fy_gpa', data=df)

plt.xlabel('SAT Score')
plt.ylabel('GPA')
plt.title('Relationship between GPA and SAT Scores')
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split


X = df.drop("fy_gpa", axis=1)
y = df["fy_gpa"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33)
```

```python
print(X_train.shape)
```

```python
print(X_test.shape)
```

```python
print(y_train.shape)
```

```python
print(y_test.shape)
```

```python
lm = LinearRegression()
model = lm.fit(X_train, y_train)
```

```python
model.predict(X_test)
```

```python
predicted_gpa = model.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, predicted_gpa)
```

The linear regression model shows an insignificant relationship between the independent and dependent variables. Evident by the r^2 score of 0.27, showing that the model is unable to consistently, accurately predict the dependent variable.
